```python
# Importing all necessary libraries of Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 1: Load the dataset
df = pd.read_csv("Combined_dataset.csv")
print(df.head())
print(f" Total Products : {df.shape[0]}")
print(f" Total Columns  : {df.shape[1]}")
```

```python
""" Step 2: Explore the Data (EDA)
EDA = Exploratory Data Analysis is always the first step before cleaning.
"""

print("Using head() : shows first 5 rows as default")
print(df.head())

print("\n Using tail() : shows last 5 rows as default")
print(df.tail())

print(f"\n Using shape()")
print(f"Rows:{df.shape[0]},Columns: {df.shape[1]}")

print("\n Viewing all column names")
for i, col in enumerate(df.columns, 1):
    print(f"{i}.{col}")

print("\n Data Types")
print(df.dtypes)

print("\n Dataset information")
df.info()
```

```python
# Step 3: Finding missing values (NaN) and using fillna/dropna to handle them

missing_values = df.isnull().sum()

print("Missing values in each column:")
print(missing_values)

# Filling missing values in important text columns

df["seller_name"] = df["seller_name"].fillna("Seller Not Mentioned")

df["seller_information"] = df["seller_information"].fillna("Information Not Available")

df["what_customers_said"] = df["what_customers_said"].fillna("No Customer Review")

print("Missing values handled successfully")

# Dropping rows where rating is missing

df = df.dropna(subset=["rating"])

print("Rows with missing ratings removed")
print("Dataset shape:", df.shape)
```

Basic operations such as selecting columns and filtering rows are performed to better understand the dataset.


```python
# Step:4 Performing basic functions like filtering rows and selecting columns

#Selecting useful columns
product_summary = df[
    ["product_id", "title", "rating", "final_price", "category"]
]
print(product_summary.head())

#Filtering products with good ratings
top_rated_products = df[df["rating"] >= 4.0]
print(top_rated_products.head())
print("Number of highly rated products:", top_rated_products.shape[0])

#Checking products from one category
print("\n Checking products from one category\n")
backpack_products = df[df["category"] == "backpacks"]
print(backpack_products.head())
```

Duplicate records are checked and removed to improve data quality and consistency.

```python
# Step 5: Removing duplicates

#Checking duplicate rows
duplicate_count = df.duplicated().sum()
print("Duplicate rows found:", duplicate_count)

#Removing duplicate rows if present
df = df.drop_duplicates()
print("Dataset shape after removing duplicates:")
print(df.shape)
```

A derived column is created to demonstrate column transformation using numerical values available in the dataset.

```python
# Step:6 Create a derived column (total_amount = price * quantity)
# Since the given dataset doesn't have quantity column i am assuming ratings_count column as quantity for practical purpose

df["total_amount"] = (
    df["initial_price"] * df["ratings_count"]
)
print(df[
    ["initial_price", "ratings_count", "total_amount"]
].head())
```

```python
# Step 7: Saving cleaned dataset

df.to_csv(
    "cleaned_shopping_dataset.csv",
    index=False
)
print("Saved Cleaned dataset")
```

```python
# Final dataset summary

print("Final rows and columns:", df.shape)

print("\nDataset Information:")
print(df.info())

print("\nNumerical Summary:")
print(df.describe())
```
